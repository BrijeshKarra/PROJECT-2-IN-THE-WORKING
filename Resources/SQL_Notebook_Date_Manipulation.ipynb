###Using DATE_TRUNC To Truncate Dates To Only Year-Month-1st Date



```python
%%sql

SELECT
DATE_TRUNC('month',orderdate)::date AS order_month,
SUM(quantity*netprice*exchangerate) AS net_revenue,
COUNT(DISTINCT customerkey) AS total_unique_customers
FROM sales

GROUP BY order_month


LIMIT 10


```

###Using TO_CHAR() To Get Only Year-Month


```python
%%sql

SELECT
TO_CHAR(orderdate,'YYYY-MM') AS order_month,
SUM(quantity*netprice*exchangerate) AS net_revenue,
COUNT(DISTINCT customerkey) AS total_unique_customers
FROM sales
GROUP BY order_month
limit 10
```

### Using DATE_PART() To Get Year, Month, And Day From A Date

```python
%%sql

SELECT
DATE_PART('year', orderdate) AS order_year,
DATE_PART('month', orderdate) AS order_month,
DATE_PART('day', orderdate) AS order_day




FROM sales
order by random()
--GROUP BY order_month
limit 10
```

### Using Extract() To Extract Year, Month, And Day From A Date

```python
%%sql

SELECT ORDERDATE,
EXTRACT(YEAR FROM orderdate) as extract_year,
EXTRACT(MONTH FROM orderdate) as extract_month,
EXTRACT(DAY FROM orderdate) as extract_day




FROM sales
order by random()
limit 10
```

```python
%%sql

SELECT
EXTRACT(YEAR FROM orderdate) AS order_year,
EXTRACT(MONTH FROM orderdate) AS order_month,

SUM(quantity*netprice*exchangerate) AS net_revenue
FROM sales
GROUP BY order_year, order_month
ORDER BY order_year, order_month
--limit 12
```

###Implementing CURRENT_DATE And Calculating Last 5 Years Revenue

```python
%%sql

SELECT

       s.orderdate,
       p.categoryname,
       SUM(s.quantity*s.netprice*s.exchangerate) AS last_5year_net_revenue

FROM sales as s
LEFT JOIN product as p ON s.productkey = p.productkey
WHERE  EXTRACT(YEAR FROM orderdate) >= EXTRACT(YEAR FROM CURRENT_DATE) -5
GROUP BY s.orderdate,p.categoryname
ORDER BY s.orderdate,p.categoryname
```

### Using INTERVAL To Get Precise Result of Last 5 Years Revenue

```python
%%sql

SELECT

       s.orderdate,
       p.categoryname,
       SUM(s.quantity*s.netprice*s.exchangerate) AS last_5year_net_revenue

FROM sales as s
LEFT JOIN product as p ON s.productkey = p.productkey
WHERE  orderdate> CURRENT_DATE - INTERVAL '5 years'
GROUP BY s.orderdate,p.categoryname
ORDER BY s.orderdate,p.categoryname
```

### Comparing Processing Times Over The Last 5 Years

```python
%%sql


SELECT
EXTRACT(YEAR FROM orderdate) as order_year,

ROUND(AVG (EXTRACT (DAYS FROM (AGE(deliverydate,orderdate) ))),2) as processing_time ,
SUM(quantity*netprice*exchangerate)::INT  net_revenue

FROM sales
WHERE orderdate >= CURRENT_DATE-INTERVAL '5 YEARS'
GROUP BY order_year

ORDER BY order_year
```
